ENSAYO TEMPRANO 201819 CON VARIABLES "PESO GRANO" Y "HUMEDAD"

```python
print("--- INICIANDO PROCESAMIENTO: ENSAYO TEMPRANO ---")

# ==============================================================================
# 0. CARGA DE DATOS 
# ==============================================================================
archivo_original = "ENSAYO_2018-2019_Tornello.xlsx"
print(f"Cargando {archivo_original} con Pandas (para cálculo) y OpenPyXL (para escritura)...")

# Leemos con Pandas para hacer la matemática
try:
    ensayo_temp1819 = pd.read_excel(archivo_original, sheet_name="ENS TEMP 201819", decimal=",")
    hum_temp1819 = pd.read_excel(archivo_original, sheet_name="HUM TEMP", decimal=",")
except Exception as e:
    print(f"Error al leer el archivo con Pandas. Asegurate de que el archivo se llama '{archivo_original}' y está en la misma carpeta.")
    raise e

# Cargamos el workbook original con openpyxl para no perder NINGÚN formato ni color
try:
    wb = load_workbook(archivo_original)
except Exception as e:
    print(f"Error al cargar el archivo con openpyxl. Detalle: {e}")
    raise e

# ==============================================================================
# 1. FUNCIÓN DE CRITERIOS AGRONÓMICOS (Tal como la definiste)
# ==============================================================================
def asignar_decision_final(grupo):
    fecha_col = None
    for c in grupo.columns:
        if c.lower() == 'fecha':
            fecha_col = c
            break
    if fecha_col is not None:
        grupo = grupo.sort_values(by=fecha_col)

    tiene_mf = grupo["MADUREZ_FISIOLOGICA"].any()
    tiene_mc = grupo["MADUREZ_COMERCIAL"].any()

    ultima_humedad = grupo["HUM"].iloc[-1]

    total_mediciones = len(grupo)
    if total_mediciones == 0:
        return "Se descarta"

    mediciones_esperadas = (grupo["ESTADO_PARCELA"] == "Esperado").sum()
    porcentaje_esperado = mediciones_esperadas / total_mediciones if total_mediciones>0 else 0

    # Umbral de 10% a 17% (Ajustado según tu indicación)
    rango_mc = (ultima_humedad >= 10.0) and (ultima_humedad <= 17.0)

    comportamiento_rescatable = porcentaje_esperado >= 0.5

    if not tiene_mf:
        return "Se descarta"
    if tiene_mf and tiene_mc:
        return "Plot Válido"
    if tiene_mf and not tiene_mc and rango_mc:
        return "Plot Válido" if comportamiento_rescatable else "Duda"
    return "Duda"

# ==============================================================================
# 2. FUNCIÓN PARA INTEGRAR RESPETANDO EL TRABAJO MANUAL (Tal como la definiste)
# ==============================================================================
def integrar_decisiones_respetando_manual(df_ensayo_orig, df_hum_procesado, nombre_ensayo="Ensayo"):
    print(f"\n--- Procesando {nombre_ensayo} ---")
    df_ens = df_ensayo_orig.copy()

    df_ens.columns = df_ens.columns.str.strip().str.upper()
    col_decision = "DECISIÓN" if "DECISIÓN" in df_ens.columns else ("DECISION" if "DECISION" in df_ens.columns else None)

    df_ens["PLOT"] = df_ens["PLOT"].astype(str).str.strip()

    decisiones_auto = df_hum_procesado.groupby("PLOT").apply(asignar_decision_final).reset_index(name="DECISION_AUTOMATICA")
    decisiones_auto["PLOT"] = decisiones_auto["PLOT"].astype(str).str.strip()

    df_final = df_ens.merge(decisiones_auto, on="PLOT", how="left")
    df_final["DECISION_AUTOMATICA"] = df_final["DECISION_AUTOMATICA"].fillna("Se descarta")

    if col_decision:
        df_final[col_decision] = df_final[col_decision].replace(r'^\s*$', np.nan, regex=True)
        hechos_a_mano = df_final[col_decision].notna().sum()
        print(f"Detectamos {hechos_a_mano} decisiones cargadas manualmente. Se conservarán.")
        df_final["DECISION_FINAL"] = df_final[col_decision].fillna(df_final["DECISION_AUTOMATICA"])
    else:
        print("No se encontró columna de decisión previa. Se aplicará lógica 100% automática.")
        df_final["DECISION_FINAL"] = df_final["DECISION_AUTOMATICA"]

    return df_final

# ==============================================================================
# 3. PREPARACIÓN DE DATOS (Fase 1 de completar vacíos en HUM TEMP)
# ==============================================================================
print("\nCompletando decisiones individuales vacías en HUM TEMP...")
hum_temp_proc = hum_temp1819.copy()

# A. Limpieza inicial: guiones '-' y espacios a NaN
cols_a_completar = ['Comportamiento Peso', 'Comportamiento HUM X', 'Decisión']
for col in cols_a_completar:
    if col in hum_temp_proc.columns:
        hum_temp_proc[col] = hum_temp_proc[col].replace(r'^[\s\-]*$', np.nan, regex=True)

# B. Ordenar y calcular diferencias
hum_temp_proc['Fecha'] = pd.to_datetime(hum_temp_proc['Fecha'], dayfirst=True)
hum_temp_proc = hum_temp_proc.sort_values(by=['Plot', 'Fecha', 'Muestreo', 'Lata'])
hum_temp_proc['HUMEDAD'] = pd.to_numeric(hum_temp_proc['HUMEDAD'], errors='coerce')
hum_temp_proc['PESO GRANO'] = pd.to_numeric(hum_temp_proc['PESO GRANO'], errors='coerce')

hum_temp_proc['DIF_HUM'] = hum_temp_proc.groupby('Plot')['HUMEDAD'].diff()
hum_temp_proc['DIF_PESO'] = hum_temp_proc.groupby('Plot')['PESO GRANO'].diff()

# C. Rellenar 'Comportamiento HUM X' y 'Comportamiento Peso' (SOLO VACÍOS)
if 'Comportamiento HUM X' in hum_temp_proc.columns:
    cond_hum_esp = (hum_temp_proc['DIF_HUM'].isna()) | (hum_temp_proc['DIF_HUM'] <= 1.0)
    mask_hum_vacia = hum_temp_proc['Comportamiento HUM X'].isna()
    hum_temp_proc.loc[mask_hum_vacia, 'Comportamiento HUM X'] = np.where(cond_hum_esp[mask_hum_vacia], 'Esperado', 'No esperado')

if 'Comportamiento Peso' in hum_temp_proc.columns:
    cond_peso_esp = (hum_temp_proc['DIF_PESO'].isna()) | (hum_temp_proc['DIF_PESO'] >= -2.0)
    mask_peso_vacia = hum_temp_proc['Comportamiento Peso'].isna()
    hum_temp_proc.loc[mask_peso_vacia, 'Comportamiento Peso'] = np.where(cond_peso_esp[mask_peso_vacia], 'Esperado', 'No esperado')

# D. Rellenar la 'Decisión' individual (SOLO VACÍOS)
if 'Decisión' in hum_temp_proc.columns:
    mask_dec_vacia = hum_temp_proc['Decisión'].isna()
    ambos_esperados = (hum_temp_proc['Comportamiento Peso'].str.strip().str.title() == 'Esperado') & \
                      (hum_temp_proc['Comportamiento HUM X'].str.strip().str.title() == 'Esperado')
    hum_temp_proc.loc[mask_dec_vacia, 'Decisión'] = np.where(ambos_esperados[mask_dec_vacia], 'Dato válido', 'Se descarta')

# ==============================================================================
# 4. ADAPTACIÓN PARA INTEGRACIÓN (Como indicaste en tu código)
# ==============================================================================
hum_temp_proc.columns = hum_temp_proc.columns.str.strip().str.upper()
if 'HUMEDAD' in hum_temp_proc.columns:
    hum_temp_proc = hum_temp_proc.rename(columns={'HUMEDAD': 'HUM'})
if 'PLOT' not in hum_temp_proc.columns:
    raise KeyError("hum_temp_proc no contiene la columna 'PLOT'.")
hum_temp_proc['PLOT'] = hum_temp_proc['PLOT'].astype(str).str.strip()

if "MADUREZ_FISIOLOGICA" not in hum_temp_proc.columns and "MF" in hum_temp_proc.columns:
    hum_temp_proc["MADUREZ_FISIOLOGICA"] = (
        hum_temp_proc["MF"].astype(str).str.strip().str.upper()
        .map({"VERDADERO": True, "FALSO": False, "TRUE": True, "FALSE": False, "1": True, "0": False})
        .fillna(False)
    )

if "MADUREZ_COMERCIAL" not in hum_temp_proc.columns and "MC" in hum_temp_proc.columns:
    hum_temp_proc["MADUREZ_COMERCIAL"] = (
        hum_temp_proc["MC"].astype(str).str.strip().str.upper()
        .map({"VERDADERO": True, "FALSO": False, "TRUE": True, "FALSE": False, "1": True, "0": False})
        .fillna(False)
    )

if "ESTADO_PARCELA" not in hum_temp_proc.columns:
    if "COMPORTAMIENTO HUM X" in hum_temp_proc.columns:
        hum_temp_proc["ESTADO_PARCELA"] = hum_temp_proc["COMPORTAMIENTO HUM X"]
    elif "COMPORTAMIENTO PESO" in hum_temp_proc.columns:
        hum_temp_proc["ESTADO_PARCELA"] = hum_temp_proc["COMPORTAMIENTO PESO"]
    else:
        hum_temp_proc["ESTADO_PARCELA"] = "No Esperado"

# Ejecutamos la integración principal
ensayo_temp1819_final = integrar_decisiones_respetando_manual(
    ensayo_temp1819,
    hum_temp_proc,
    "Ensayo Temprano"
)

print("\n✅ ¡Listo! Decisiones combinadas exitosamente.")
print("\nDistribución final Temprano:")
print(ensayo_temp1819_final["DECISION_FINAL"].value_counts())


# ==============================================================================
# 5. ESCRITURA SEGURA CON OPENPYXL (Mantiene Formato, Colores y Choices)
# ==============================================================================
archivo_salida = "ENSAYO_TEMPRANO_FINAL.xlsx"
print(f"\nInyectando datos en formato original y guardando como: {archivo_salida}...")

def get_col_idx(ws, col_name):
    """Busca en la primera fila el número de la columna deseada ignorando mayúsculas/espacios"""
    for cell in ws[1]:
        if cell.value and str(cell.value).strip().upper() == col_name.strip().upper():
            return cell.column
    return None

try:
    # --- A. Actualizar Hoja HUM TEMP ---
    ws_hum = wb["HUM TEMP"]
    # Volvemos a las columnas originales (con mayúsculas/minúsculas) para buscar en openpyxl
    col_cp = get_col_idx(ws_hum, 'Comportamiento Peso')
    col_ch = get_col_idx(ws_hum, 'Comportamiento HUM X')
    col_dec_hum = get_col_idx(ws_hum, 'Decisión')
    
    # Usamos enumerate para tener un contador entero seguro en lugar de depender del índice de Pandas
    for i, (idx, row) in enumerate(hum_temp1819.iterrows()):
        excel_row = i + 2 # +2 porque enumerate empieza en 0 y Excel en 1 + 1 del encabezado
        
        # Obtenemos los valores calculados del dataframe procesado
        val_cp = hum_temp_proc.loc[idx, 'COMPORTAMIENTO PESO'] if 'COMPORTAMIENTO PESO' in hum_temp_proc.columns else None
        val_ch = hum_temp_proc.loc[idx, 'COMPORTAMIENTO HUM X'] if 'COMPORTAMIENTO HUM X' in hum_temp_proc.columns else None
        val_dec = hum_temp_proc.loc[idx, 'DECISIÓN'] if 'DECISIÓN' in hum_temp_proc.columns else (hum_temp_proc.loc[idx, 'DECISION'] if 'DECISION' in hum_temp_proc.columns else None)
        
        # Solo escribimos si no es NaN (respetamos lo que ya estaba o escribimos lo nuevo)
        if col_cp and pd.notna(val_cp): ws_hum.cell(row=excel_row, column=col_cp).value = val_cp
        if col_ch and pd.notna(val_ch): ws_hum.cell(row=excel_row, column=col_ch).value = val_ch
        if col_dec_hum and pd.notna(val_dec): ws_hum.cell(row=excel_row, column=col_dec_hum).value = val_dec

    # --- B. Actualizar Hoja ENS TEMP 201819 ---
    ws_ens = wb["ENS TEMP 201819"]
    # Buscamos la columna de decisión en la hoja principal
    col_dec_ens = get_col_idx(ws_ens, 'Decisión')
    if not col_dec_ens:
         col_dec_ens = get_col_idx(ws_ens, 'DECISION')
    
    # Inyectamos DECISION_FINAL
    for i, (idx, row) in enumerate(ensayo_temp1819.iterrows()):
        excel_row = i + 2
        val_final = ensayo_temp1819_final.loc[idx, 'DECISION_FINAL']
        if col_dec_ens and pd.notna(val_final):
             ws_ens.cell(row=excel_row, column=col_dec_ens).value = val_final

    # Guardamos el workbook con TODO el formato intacto
    wb.save(archivo_salida)
    print(f"✅ Exportación exitosa. Archivo creado: {archivo_salida}")
    print("¡Tu formato y los colores del co-director están a salvo!")

except Exception as e:
    print(f"❌ Error al guardar el archivo con openpyxl: {e}")
    print("Asegurate de tener el archivo cerrado en Excel antes de correr el script.")
```

```python
print("--- INICIANDO PROCESAMIENTO: ENSAYO TEMPRANO ---")

# ======================================================================
# FASE 1: COMPLETAR FILAS INDIVIDUALES FALTANTES (HUM TEMP)
# ======================================================================
hum_temp1819_procesado = hum_temp1819.copy()
ensayo_temp_proc = ensayo_temp1819.copy()

# 1. Limpiar espacios o guiones a NaN para detectarlos
cols_a_completar = ['Comportamiento Peso', 'Comportamiento HUM X', 'Decisión']
for col in cols_a_completar:
    if col in hum_temp1819_procesado.columns:
        hum_temp1819_procesado[col] = hum_temp1819_procesado[col].replace(r'^[\s\-]*$', np.nan, regex=True)

# 2. Aseguramos tipo fecha y orden cronológico
hum_temp1819_procesado['Fecha'] = pd.to_datetime(hum_temp1819_procesado['Fecha'], dayfirst=True)
hum_temp1819_procesado = hum_temp1819_procesado.sort_values(by=['Plot', 'Fecha', 'Muestreo', 'Lata'])

# 3. Forzar numéricos para evitar errores matemáticos
hum_temp1819_procesado['HUMEDAD'] = pd.to_numeric(hum_temp1819_procesado['HUMEDAD'], errors='coerce')
hum_temp1819_procesado['PESO GRANO'] = pd.to_numeric(hum_temp1819_procesado['PESO GRANO'], errors='coerce')

hum_temp1819_procesado['DIF_HUM'] = hum_temp1819_procesado.groupby('Plot')['HUMEDAD'].diff()
hum_temp1819_procesado['DIF_PESO'] = hum_temp1819_procesado.groupby('Plot')['PESO GRANO'].diff()

# 4. Rellenar vacíos de Comportamiento usando .loc
if 'Comportamiento HUM X' in hum_temp1819_procesado.columns:
    cond_hum_esp = (hum_temp1819_procesado['DIF_HUM'].isna()) | (hum_temp1819_procesado['DIF_HUM'] <= 1.0)
    mask_hum = hum_temp1819_procesado['Comportamiento HUM X'].isna()
    hum_temp1819_procesado.loc[mask_hum, 'Comportamiento HUM X'] = np.where(cond_hum_esp[mask_hum], 'Esperado', 'No esperado')

if 'Comportamiento Peso' in hum_temp1819_procesado.columns:
    cond_peso_esp = (hum_temp1819_procesado['DIF_PESO'].isna()) | (hum_temp1819_procesado['DIF_PESO'] >= -2.0)
    mask_peso = hum_temp1819_procesado['Comportamiento Peso'].isna()
    hum_temp1819_procesado.loc[mask_peso, 'Comportamiento Peso'] = np.where(cond_peso_esp[mask_peso], 'Esperado', 'No esperado')

# 5. Rellenar vacíos de Decisión de la fila individual
if 'Decisión' in hum_temp1819_procesado.columns:
    # Estandarizamos para evitar problemas de mayúsculas ("Esperado" vs "esperado")
    ambos_esperados = (hum_temp1819_procesado['Comportamiento Peso'].str.strip().str.title() == 'Esperado') & \
                      (hum_temp1819_procesado['Comportamiento HUM X'].str.strip().str.title() == 'Esperado')
    mask_dec = hum_temp1819_procesado['Decisión'].isna()
    hum_temp1819_procesado.loc[mask_dec, 'Decisión'] = np.where(ambos_esperados[mask_dec], 'Dato válido', 'Se descarta')
```

```python
# ======================================================================
# FASE 2: FUNCIÓN DE CRITERIOS AGRONÓMICOS ROBUSTA (TEMPRANO)
# ======================================================================
def asignar_decision_final_temprano(grupo):
    if 'Fecha' in grupo.columns:
        grupo = grupo.sort_values(by='Fecha')
        
    # Detectamos MF y MC adaptando a cómo están escritas en Temprano (Verdadero, True, 1)
    tiene_mf = False
    if 'MF' in grupo.columns:
        tiene_mf = grupo['MF'].astype(str).str.strip().str.upper().isin(['TRUE', 'VERDADERO', '1', '1.0']).any()
    
    tiene_mc = False
    if 'MC' in grupo.columns:
        tiene_mc = grupo['MC'].astype(str).str.strip().str.upper().isin(['TRUE', 'VERDADERO', '1', '1.0']).any()
        
    total_mediciones = len(grupo)
    if total_mediciones == 0:
        return "Se descarta"
        
    ultima_humedad = grupo['HUMEDAD'].iloc[-1]
    
    # En temprano, evaluamos el estado contando las filas que son 'Dato válido'
    mediciones_esperadas = (grupo['Decisión'].str.strip().str.capitalize() == 'Dato válido').sum()
    porcentaje_esperado = mediciones_esperadas / total_mediciones if total_mediciones > 0 else 0
    
    UMBRAL_CERCA_MC = 17.0
    cerca_de_mc = ultima_humedad <= UMBRAL_CERCA_MC
    comportamiento_rescatable = porcentaje_esperado >= 0.5
    
    # --- Reglas de decisión (Adaptadas de tu código robusto) ---
    if not tiene_mf:
        return "Se descarta"
    if tiene_mf and tiene_mc:
        return "Plot Válido"
    if tiene_mf and not tiene_mc and cerca_de_mc:
        return "Plot Válido" if comportamiento_rescatable else "Duda"
        
    # El "else 'Duda' no lo sigas": si no cumple nada de lo anterior, se descarta el dato
    return "Se descarta"

# Aplicamos la función a nivel Plot
decisiones_auto_temprano = hum_temp1819_procesado.groupby('Plot').apply(asignar_decision_final_temprano).reset_index(name='DECISION_AUTO')
```

```python
# ======================================================================
# FASE 3: INTEGRAR RESPETANDO EL TRABAJO MANUAL (ENSAYO TEMPRANO)
# ======================================================================
def integrar_decisiones_respetando_manual(df_ensayo_orig, df_decisiones_auto):
    df_ens = df_ensayo_orig.copy()
    
    # Estandarizamos PLOT para cruzar correctamente
    df_ens['PLOT'] = df_ens['PLOT'].astype(str).str.strip()
    df_decisiones_auto['Plot'] = df_decisiones_auto['Plot'].astype(str).str.strip()
    
    # Limpiamos columna manual original (guiones/espacios a NaN)
    col_decision = 'Decisión' if 'Decisión' in df_ens.columns else ('DECISION' if 'DECISION' in df_ens.columns else None)
    if col_decision:
        df_ens[col_decision] = df_ens[col_decision].replace(r'^[\s\-]*$', np.nan, regex=True)
    
    # Cruzamos ambas tablas
    df_final = df_ens.merge(df_decisiones_auto, left_on='PLOT', right_on='Plot', how='left')
    
    # Rellenamos sólo los vacíos, respetando tu trabajo manual
    if col_decision:
        mask_vacia = df_final[col_decision].isna()
        hechos_a_mano = (~mask_vacia).sum()
        print(f"Detectamos {hechos_a_mano} decisiones de Plot cargadas manualmente. ¡Se conservarán intactas!")
        
        df_final.loc[mask_vacia, col_decision] = df_final.loc[mask_vacia, 'DECISION_AUTO']
        # Si alguno quedó vacío porque no estaba en la hoja de humedad, lo descartamos por defecto
        df_final[col_decision] = df_final[col_decision].fillna("Se descarta")
    else:
        df_final['Decisión'] = df_final['DECISION_AUTO'].fillna("Se descarta")
        
    # Limpieza de columnas temporales
    df_final = df_final.drop(columns=['DECISION_AUTO', 'Plot'], errors='ignore')
    return df_final

ensayo_temp1819_final = integrar_decisiones_respetando_manual(ensayo_temp1819, decisiones_auto_temprano)

# Resultados
print("\n✅ Ensayo Temprano procesado completamente.")
col_decision_final = 'Decisión' if 'Decisión' in ensayo_temp1819_final.columns else 'DECISION'
print("\n--- Distribución Final de Decisiones (A nivel Parcela) ---")
print(ensayo_temp1819_final[col_decision_final].value_counts(dropna=False))
```

```python
# 1. FUNCIÓN DE CRITERIOS AGRONÓMICOS
def asignar_decision_final(grupo):
    # Aseguramos que el grupo esté ordenado cronológicamente por la columna de fecha (si existe)
    fecha_col = None
    for c in grupo.columns:
        if c.lower() == 'fecha':
            fecha_col = c
            break
    if fecha_col is not None:
        grupo = grupo.sort_values(by=fecha_col)

    # Detectamos si existe Madurez Fisiológica (MF) o Comercial (MC) en el grupo
    tiene_mf = grupo["MADUREZ_FISIOLOGICA"].any()
    tiene_mc = grupo["MADUREZ_COMERCIAL"].any()

    # Tomamos la última medición de humedad (última fila cronológica)
    ultima_humedad = grupo["HUM"].iloc[-1]

    # Validaciones básicas
    total_mediciones = len(grupo)
    if total_mediciones == 0:
        return "Se descarta"

    # Porcentaje de mediciones que siguen el comportamiento 'Esperado'
    mediciones_esperadas = (grupo["ESTADO_PARCELA"] == "Esperado").sum()
    porcentaje_esperado = mediciones_esperadas / total_mediciones if total_mediciones>0 else 0

    # Umbral usado para considerar que la parcela está 'cerca' de la MC.
    # Se dejó en 17% porque 20% se consideró alto para extrapolar la MC (ajustable).
    UMBRAL_CERCA_MC = 17.0
    cerca_de_mc = ultima_humedad <= UMBRAL_CERCA_MC

    # Si la mayoría de las mediciones son 'Esperado', consideramos rescatable
    comportamiento_rescatable = porcentaje_esperado >= 0.5

    # Reglas de decisión (explicadas):
    # - Si no hay MF, no podemos determinar correctamente → 'Se descarta'.
    # - Si hay MF y MC observada → 'Plot Válido'.
    # - Si hay MF pero no MC observada: si la última humedad está cerca
    #   y el comportamiento es rescatable, lo aceptamos; si no, queda 'Duda'.
    # - En cualquier otro caso devolvemos 'Duda'.
    if not tiene_mf:
        return "Se descarta"
    if tiene_mf and tiene_mc:
        return "Plot Válido"
    if tiene_mf and not tiene_mc and cerca_de_mc:
        return "Plot Válido" if comportamiento_rescatable else "Duda"
    return "Duda"

# 2. FUNCIÓN PARA INTEGRAR RESPETANDO EL TRABAJO MANUAL
def integrar_decisiones_respetando_manual(df_ensayo_orig, df_hum_procesado, nombre_ensayo="Ensayo"):
    print(f"\n--- Procesando {nombre_ensayo} ---")
    df_ens = df_ensayo_orig.copy()

    # Normalizamos nombres de columnas para comparar de forma segura
    df_ens.columns = df_ens.columns.str.strip().str.upper()
    # Buscamos columna de decisión manual (puede tener tilde o no)
    col_decision = "DECISIÓN" if "DECISIÓN" in df_ens.columns else ("DECISION" if "DECISION" in df_ens.columns else None)

    # Normalizamos PLOT como string y sin espacios
    df_ens["PLOT"] = df_ens["PLOT"].astype(str).str.strip()

    # A) Calculamos la decisión automática para TODOS los plots a partir de los datos de humedad
    # Asegurarse de que df_hum_procesado tiene la columna 'PLOT' y las columnas que usa asignar_decision_final
    decisiones_auto = df_hum_procesado.groupby("PLOT").apply(asignar_decision_final).reset_index(name="DECISION_AUTOMATICA")
    decisiones_auto["PLOT"] = decisiones_auto["PLOT"].astype(str).str.strip()

    # B) Cruzamos los datos con la planilla del ensayo (left join para conservar todos los plots en la planilla)
    df_final = df_ens.merge(decisiones_auto, on="PLOT", how="left")
    # Si faltan datos de humedad para un plot, asumimos 'Se descarta' por defecto
    df_final["DECISION_AUTOMATICA"] = df_final["DECISION_AUTOMATICA"].fillna("Se descarta")

    # C) Respetar lo manual: si ya hay una decisión cargada manualmente, conservarla;
    #    si está vacía, rellenarla con lo calculado por la máquina.
    if col_decision:
        # Convertimos celdas que solo contienen espacios en NaN para identificarlas como vacías
        df_final[col_decision] = df_final[col_decision].replace(r'^\s*$', np.nan, regex=True)

        # Informamos cuántas decisiones estaban cargadas manualmente
        hechos_a_mano = df_final[col_decision].notna().sum()
        print(f"Detectamos {hechos_a_mano} decisiones cargadas manualmente. Se conservarán.")

        # Creamos la columna final: preferimos la decisión manual si existe, sino la automática
        df_final["DECISION_FINAL"] = df_final[col_decision].fillna(df_final["DECISION_AUTOMATICA"])
    else:
        print("No se encontró columna de decisión previa. Se aplicará lógica 100% automática.")
        df_final["DECISION_FINAL"] = df_final["DECISION_AUTOMATICA"]

    return df_final

# =====================================================================
# 3. EJECUCIÓN PARA AMBOS ENSAYOS (ejemplo de uso)
# Asegurarse de pasar las variables procesadas correctas (hum_temp1819, hum_tardio1819_procesado, etc.)

# Normalizar y preparar los dataframes de humedad antes de integrarlos (evita KeyError: 'PLOT')
# 1) Ensayo Temprano
hum_temp_proc = hum_temp1819.copy()
hum_temp_proc.columns = hum_temp_proc.columns.str.strip().str.upper()
if 'HUMEDAD' in hum_temp_proc.columns:
    hum_temp_proc = hum_temp_proc.rename(columns={'HUMEDAD': 'HUM'})
if 'PLOT' not in hum_temp_proc.columns:
    raise KeyError("hum_temp1819 no contiene la columna 'PLOT' (después de normalizar).")
hum_temp_proc['PLOT'] = hum_temp_proc['PLOT'].astype(str).str.strip()

# Adaptamos hum_temp_proc al esquema esperado por asignar_decision_final
hum_temp_proc = hum_temp_proc.copy()

if "HUM" not in hum_temp_proc.columns and "HUMEDAD" in hum_temp_proc.columns:
    hum_temp_proc["HUM"] = hum_temp_proc["HUMEDAD"]

if "MADUREZ_FISIOLOGICA" not in hum_temp_proc.columns and "MF" in hum_temp_proc.columns:
    hum_temp_proc["MADUREZ_FISIOLOGICA"] = (
        hum_temp_proc["MF"].astype(str).str.strip().str.upper()
        .map({"VERDADERO": True, "FALSO": False, "TRUE": True, "FALSE": False})
        .fillna(False)
    )

if "MADUREZ_COMERCIAL" not in hum_temp_proc.columns and "MC" in hum_temp_proc.columns:
    hum_temp_proc["MADUREZ_COMERCIAL"] = (
        hum_temp_proc["MC"].astype(str).str.strip().str.upper()
        .map({"VERDADERO": True, "FALSO": False, "TRUE": True, "FALSE": False})
        .fillna(False)
    )

if "ESTADO_PARCELA" not in hum_temp_proc.columns:
    if "Comportamiento HUM X" in hum_temp_proc.columns:
        hum_temp_proc["ESTADO_PARCELA"] = hum_temp_proc["Comportamiento HUM X"]
    elif "Comportamiento Peso" in hum_temp_proc.columns:
        hum_temp_proc["ESTADO_PARCELA"] = hum_temp_proc["Comportamiento Peso"]
    else:
        hum_temp_proc["ESTADO_PARCELA"] = "No Esperado"

ensayo_temp1819_final = integrar_decisiones_respetando_manual(
    ensayo_temp1819,
    hum_temp_proc,
    "Ensayo Temprano"
)

print("\n✅ ¡Listo! Decisiones combinadas exitosamente.")
print("\nDistribución final Temprano:")
print(ensayo_temp1819_final["DECISION_FINAL"].value_counts())
```

```python
df = pd.read_excel("ENSAYO_2018-2019_GoogleSheets.xlsx", sheet_name="ENS TEMP 201819", decimal=",")
ensayo_tardio1819 = pd.read_excel("ENSAYO_2018-2019_GoogleSheets.xlsx", sheet_name="ENS TARDIO 201819", decimal=",")

def clasificar_filas(df):
    
    df = df.sort_values("Fecha")
    df["Esperado"] = True
    
    for i in range(1, len(df)):
        
        hum_actual = df.iloc[i]["HUM"]
        hum_prev = df.iloc[i-1]["HUM"]
        
        ps_actual = df.iloc[i]["Peso_Seco"]
        ps_prev = df.iloc[i-1]["Peso_Seco"]
        
        if hum_actual > hum_prev or ps_actual < ps_prev:
            df.iloc[i, df.columns.get_loc("Esperado")] = False
            
    return df


def calcular_MF(df):
    
    max_ps = df["Peso_Seco"].max()
    
    candidatos = df[
        (df["HUM"] >= 32) &
        (df["HUM"] <= 40) &
        (df["Peso_Seco"] == max_ps)
    ]
    
    if len(candidatos) > 0:
        return candidatos.iloc[0]["Fecha"], True
    
    return None, False


def extrapolar_MC(df):

    df = df.sort_values("Fecha")

    x = (df["Fecha"] - df["Fecha"].min()).dt.days
    y = df["HUM"]

    coef = np.polyfit(x, y, 1)

    a, b = coef
    
    x_target = (14.5 - b) / a
    
    fecha_mc = df["Fecha"].min() + pd.Timedelta(days=x_target)

    return fecha_mc


def calcular_MC(df):

    rango = df[(df["HUM"] >= 10) & (df["HUM"] <= 16)]

    if len(rango) > 0:
        return rango.iloc[0]["Fecha"], "observada"

    ultima = df.iloc[-1]

    if ultima["HUM"] > 14.5:
        return extrapolar_MC(df), "extrapolada"

    return None, "duda"


def analizar_plot(df_plot):

    df_plot = clasificar_filas(df_plot)

    df_validas = df_plot[df_plot["Esperado"] == True]

    if len(df_validas) == 0:
        return {"decision": "Se descarta"}

    MF, MF_valida = calcular_MF(df_validas)

    if MF_valida == False:
        return {"decision": "Se descarta"}

    MC, tipo = calcular_MC(df_validas)

    if MC is None:
        return {"decision": "Duda", "MF": MF}

    return {
        "decision": "Plot válido",
        "MF": MF,
        "MC": MC,
        "tipo_MC": tipo
    }
```

- La humedad (HUM) debe disminuir en sucesivos muestreos.
- El peso seco (Peso Seco) debe aumentar en sucesivos muestreos.

"No esperado":

```python
# Limpiar nombres de columnas
df_gdu_temp.columns = df_gdu_temp.columns.str.strip()
df_gdu_tardio.columns = df_gdu_tardio.columns.str.strip()

# Verificar que la columna 'HUM' exista en ambos DataFrames
if "HUM" not in df_gdu_temp.columns or "HUM" not in df_gdu_tardio.columns:
    raise KeyError("La columna 'HUM' no se encuentra en uno o ambos DataFrames. Verifique los nombres de las columnas.")

# Ordenar para evaluar correctamente el comportamiento por fecha
df_gdu_temp = df_gdu_temp.sort_values(by=["Plot", "Fecha", "Muestreo", "Lata"])
df_gdu_tardio = df_gdu_tardio.sort_values(by=["Plot", "Fecha", "Muestreo", "Lata"])

# Función para evaluar comportamiento en cada parcela
def evaluar_plot(grupo):
    humedad_decreciente = grupo["HUM"].is_monotonic_decreasing
    peso_creciente = grupo["Peso Seco"].is_monotonic_increasing
    return "Esperado" if humedad_decreciente and peso_creciente else "No esperado"

# Agrupar por `Plot`, `Fecha`, `Muestreo`, `Lata`
df_gdu_temp["Estado_Parcela"] = df_gdu_temp.groupby(["Plot", "Fecha", "Muestreo", "Lata"]).apply(
    lambda grupo: evaluar_plot(grupo)
).reset_index(level=[0, 1, 2, 3], drop=True)

df_gdu_tardio["Estado_Parcela"] = df_gdu_tardio.groupby(["Plot", "Fecha", "Muestreo", "Lata"]).apply(
    lambda grupo: evaluar_plot(grupo)
).reset_index(level=[0, 1, 2, 3], drop=True)

# Guardar en un nuevo archivo
with pd.ExcelWriter("ENSAYO_2018-19_LOGICA.xlsx", engine="xlsxwriter") as writer:
    df_gdu_temp.to_excel(writer, sheet_name="HUM TEMP", index=False)
    df_gdu_tardio.to_excel(writer, sheet_name="HUM TARD", index=False)

print("✅ Clasificación corregida y aplicada a todas las filas correctamente.")
```

```python
import pandas as pd

def evaluar_comportamiento(df):
    df = df.sort_values(by='Fecha')  # Ordenar por fecha para evaluar tendencias
    df['Comportamiento HM X'] = 'Esperado'
    df['Comportamiento Peso'] = 'Esperado'
    
    for i in range(1, len(df)):
        if df.iloc[i]['HUM'] > df.iloc[i-1]['HUM']:
            df.at[df.index[i], 'Comportamiento HM X'] = 'No Esperado'
        if df.iloc[i]['Peso Seco'] < df.iloc[i-1]['Peso Seco']:
            df.at[df.index[i], 'Comportamiento Peso'] = 'No Esperado'
    
    return df

def determinar_mf(df):
    df_mf = df[(df['HUM'] >= 32) & (df['HUM'] <= 40)]
    if not df_mf.empty:
        mf = df_mf.loc[df_mf['Peso Seco'].idxmax(), 'Fecha']  # Mayor peso seco dentro del rango
        return mf
    return None

def calcular_tsg(df, mf):
    df_mc = df[(df['HUM'] >= 10) & (df['HUM'] <= 16)]
    if not df_mc.empty:
        mc = df_mc.iloc[0]['Fecha']  # Primera fecha con humedad en rango MC
        hum_mf = df.loc[df['Fecha'] == mf, 'HUM'].values[0]
        hum_mc = df.loc[df['Fecha'] == mc, 'HUM'].values[0]
        tsg = (hum_mf - hum_mc) / (mc - mf).days if (mc - mf).days > 0 else None
        return mc, tsg
    return None, None

def procesar_plots(df):
    df.columns = df.columns.str.upper().str.strip()  # Convertir nombres de columnas a mayúsculas
    
    if 'PLOT' not in df.columns:
        raise KeyError("La columna 'PLOT' no está en el DataFrame. Verifique los datos de entrada.")

    df['PLOT'] = df['PLOT'].astype(int)  # Asegurar que PLOT es numérico
    df['Comportamiento Peso'] = df['Comportamiento Peso'].astype(str)  # Asegurar texto
    df['Decisión'] = df['Decisión'].astype(str)  # Asegurar texto
    df['MF'] = df['MF'].astype(bool)  # MF como VERDADERO/FALSO
    df['MC'] = df['MC'].astype(bool)  # MC como VERDADERO/FALSO
    df['TSG'] = pd.to_numeric(df['TSG'], errors='coerce')  # TSG numérico
    
    for plot in df['PLOT'].unique():
        subset = df[df['PLOT'] == plot].copy()
        subset = evaluar_comportamiento(subset)
        mf = determinar_mf(subset)
        
        if mf:
            mc, tsg = calcular_tsg(subset, mf)
            decision = 'Plot Válido'
        else:
            mc, tsg, decision = None, None, 'Se descarta'
        
        df.loc[df['PLOT'] == plot, 'Decisión'] = decision
        
        if mc:
            df.loc[(df['PLOT'] == plot) & (df['MC'] == True), 'TSG'] = tsg  # Solo asignar en filas con MC VERDADERO
    
    return df

# Cargar el DataFrame con datos reales
hum_tard = pd.read_excel("ENSAYO_2018-19_GDUs.xlsx", sheet_name="HUM TARD", decimal=",")

# Convertir nombres de columnas a mayúsculas para evitar problemas
hum_tard.columns = hum_tard.columns.str.upper().str.strip()

# Procesar los datos
hum_tard_procesado = procesar_plots(hum_tard)

# Guardar el archivo corregido
hum_tard_procesado.to_excel("HUM_TARD_PROCESADO.xlsx", index=False, decimal=",")
print("Proceso completado. Archivo guardado como 'HUM_TARD_PROCESADO.xlsx'.")
```

```python
import pandas as pd

# Cargar los datos de los ensayos
ens_temprano = pd.read_excel("ENSAYO_2018-2019_Tornello.xlsx", sheet_name="ENS TEMP 201819", decimal=",")
ens_tardio = pd.read_excel("ENSAYO_2018-2019_Tornello.xlsx", sheet_name="ENS TARDIO 201819", decimal=",")

# Normalizar nombres de columnas (convertir a mayúsculas y quitar espacios)
ens_temprano.columns = ens_temprano.columns.str.upper().str.strip()
ens_tardio.columns = ens_tardio.columns.str.upper().str.strip()

# Seleccionar columnas de interés y renombrar si es necesario
columnas_deseadas = [
    "AÑO", "LOCALIDAD", "ENSAYO", "PLOT", "BLOC", "IBLK", "PEDIGREE",
    "GDU_VT", "GDU_R1", "GDU_ASI", "GDU_MF", "FECHA MF", "GDU_MC", "FECHA MC",
    "TASA DE SECADO DEL GRANO", "PESO SECO MIL", "CHALAS"
]

# Crear estructura base con valores fijos
def preparar_dataframe(df, ensayo):
    df = df.copy()
    df["AÑO"] = df["PLOT"].apply(lambda x: "2018" if int(x) < 200 else "2019")  # Definir año según criterio
    df["LOCALIDAD"] = "Pergamino"  # Localidad fija
    df["ENSAYO"] = ensayo  # Tipo de ensayo ('Temprano' o 'Tardío')

    # Asegurar que todas las columnas existen en el DataFrame, aunque estén vacías
    for col in columnas_deseadas:
        if col not in df.columns:
            df[col] = None  # Crear columna vacía si no existe

    return df[columnas_deseadas]  # Devolver solo las columnas deseadas en el orden correcto

# Procesar ambos ensayos
df_temprano = preparar_dataframe(ens_temprano, "Temprano")
df_tardio = preparar_dataframe(ens_tardio, "Tardío")

# Unir ambos ensayos en un solo DataFrame
df_final = pd.concat([df_temprano, df_tardio], ignore_index=True)

# Guardar en un archivo Excel
df_final.to_excel("BASE_DATOS_UNIFICADA.xlsx", index=False)

print("✅ Proceso completado. Archivo guardado como 'BASE_DATOS_UNIFICADA.xlsx'.")
```
